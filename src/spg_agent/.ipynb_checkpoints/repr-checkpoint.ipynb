```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('text', usetex=True) # https://ercanozturk.org/2017/12/16/python-matplotlib-plots-in-latex/
plt.rc('font', family='serif')
```

```python
data = ['repr-exp-1-dqn-comb.npy',
        'repr-exp-1-dqn.npy',
        'repr-exp-1-spg.npy',            
        'repr-exp-2-dqn-comb.npy',
        'repr-exp-2-dqn.npy',
        'repr-exp-2-spg.npy',
        'repr-exp-3-dqn-comb.npy',
        'repr-exp-3-dqn.npy',
        'repr-exp-3-spg.npy']

experiment_subtitles = ['Rewarding index \n on a single specified query',
                        'Rewarding indicies \n on several specified queries',
                        'Rewarding index \n on single specified query \n given higher dimensional context']

fig = plt.figure(figsize=(27,18))

for experiment_idx in range(3):
    
    for agent_idx in range(3):
        
        idx = experiment_idx*3 + agent_idx
        runs = np.load(data[idx])
        
        ax = fig.add_subplot(3, 3, idx+1)  # i rows, j columns, slow 
        
        for run_idx in range(runs.shape[0]):
    
            ax.plot(runs[run_idx,:,0], runs[run_idx,:,1])
        
        plt.ylim([-1.1,1.1])
        #plt.xlabel('steps', fontsize=18)
        #plt.ylabel('running average reward', fontsize=18)
        plt.tick_params(axis='both', labelsize=24)
        
        
        # cols
        if agent_idx == 0:
            tags = ['(a)', '(b)', '(c)']
            ax.set_ylabel(tags[experiment_idx], fontsize=36, rotation=90)
        
        # rows
        if experiment_idx == 0:
            agents = ['DQN (combinatorial action schema)', 'DQN', 'SPG']
            ax.set_title(agents[agent_idx], fontsize=36, 
                         pad = plt.rcParams['axes.titlepad'] + 5)
    
        if experiment_idx == 0:
            time_per_step = round(np.mean(runs[:,-1,2] / runs[:,-1,0]), 4)
                                                    
            ax.text(0.05, 0.85, "{} sec per step".format(time_per_step), 
                    fontsize=24, transform=ax.transAxes,
                    bbox=dict(facecolor='red', alpha=0.25))
            
        if experiment_idx==2 and agent_idx==1:
            ax.set_xlabel('steps', fontsize=24)

        ax.hlines(1,xmin=0,xmax=100000, alpha=0.5, color='red', linewidth=.5)

fig.suptitle("Training trajectories as running average rewards \n"
             "Rewards per step in \{-1,1\} \n"
             "Reward index on (a) a single specified query (b) several specified queries (c) a single specified query given context",
             x=0.5, y=1.1,
            fontsize=36)


plt.tight_layout()
```

```python
# NB! DQN don't require 100,000 steps. seems like learning ability stems from exploration-exploitation strategy, not # of steps.
```
